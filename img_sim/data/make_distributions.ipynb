```python
lens_df2 = pd.read_csv("../../../lp2/lenses_DESc.txt", 
                    sep="\s+",
                    header=None,
                    usecols=cols_lens,
                    names=names
                   )

lens_df3 = pd.read_csv("../../../lp3/lenses_DESc.txt", 
                    sep="\s+", 
                    header=None,
                    usecols=cols_lens,
                    names=names
                   )

lens_df4 = pd.read_csv("../../../lp4/lenses_DESc.txt", 
                    sep="\s+", 
                    header=None,
                    usecols=cols_lens,
                    names=names
                   )
lens_df5 = pd.read_csv("../../../lp5/lenses_DESc.txt", 
                    sep="\s+", 
                    header=None,
                    usecols=cols_lens,
                    names=names
                   )
lens_df6 = pd.read_csv("../../../lp6/lenses_DESc.txt", 
                    sep="\s+", 
                    header=None,
                    usecols=cols_lens,
                    names=names
                   )
lens_df7 = pd.read_csv("../../../lp7/lenses_DESc.txt", 
                    sep="\s+", 
                    header=None,
                    usecols=cols_lens,
                    names=names
                   )
lens_df8 = pd.read_csv("../../../lp8/lenses_DESc.txt", 
                    sep="\s+", 
                    header=None,
                    usecols=cols_lens,
                    names=names
                   )
lens_df8.shape
```

```python
source_df2 = pd.read_csv("../../../lp2/lenses_DESc.txt", 
                    sep="\s+",
                    header=None,
                    usecols=cols_source,
                    names=names
                   )

source_df3 = pd.read_csv("../../../lp3/lenses_DESc.txt", 
                    sep="\s+", 
                    header=None,
                    usecols=cols_source,
                    names=names
                   )

source_df4 = pd.read_csv("../../../lp4/lenses_DESc.txt", 
                    sep="\s+", 
                    header=None,
                    usecols=cols_source,
                    names=names
                   )
source_df5 = pd.read_csv("../../../lp5/lenses_DESc.txt", 
                    sep="\s+", 
                    header=None,
                    usecols=cols_source,
                    names=names
                   )
source_df6 = pd.read_csv("../../../lp6/lenses_DESc.txt", 
                    sep="\s+", 
                    header=None,
                    usecols=cols_source,
                    names=names
                   )
source_df7 = pd.read_csv("../../../lp7/lenses_DESc.txt", 
                    sep="\s+", 
                    header=None,
                    usecols=cols_source,
                    names=names
                   )
source_df8 = pd.read_csv("../../../lp8/lenses_DESc.txt", 
                    sep="\s+", 
                    header=None,
                    usecols=cols_source,
                    names=names
                   )
source_df8.shape
```

```python
full_df = pd.concat([lens_df1,
                     lens_df2,
                     lens_df3,
                     lens_df4,
                     lens_df5,
                     lens_df6,
                     lens_df7,
                     lens_df8,
                     source_df1,
                     source_df2,
                     source_df3,
                     source_df4,
                     source_df5,
                     source_df6,
                     source_df7,
                     source_df8,                     
                     ], 
                     axis=0)

full_df.shape
```

```python
for param in full_df.columns:
    fig = plt.figure()
    plt.hist(full_df[param], bins=99)
    plt.suptitle(param)
    plt.grid()
```

```python
print(full_df.shape)

full_df_cut = full_df[(full_df['z'] > 0) & (full_df['z'] < 0.7) &
                (full_df['mag_g'] > 0) & (full_df['mag_g'] < 22) &
                (full_df['mag_r'] > 0) & (full_df['mag_r'] < 20) &
                (full_df['mag_i'] > 0) & (full_df['mag_i'] < 19) &
                (full_df['mag_z'] > 0) & (full_df['mag_z'] < 19) 
                ]

print(full_df_cut.shape)
```

```python
fig, axs = plt.subplots(1, 5, figsize=(15,3))
axs[0].set_title('g')
axs[0].hist(full_df_cut['mag_g'], bins=50)
axs[0].grid(alpha=0.6)

axs[1].set_title('r')
axs[1].hist(full_df_cut['mag_r'], bins=50)
axs[1].grid(alpha=0.6)

axs[2].set_title('i')
axs[2].hist(full_df_cut['mag_i'], bins=50)
axs[2].grid(alpha=0.6)

axs[3].set_title('z')
axs[3].hist(full_df_cut['mag_z'], bins=50)
axs[3].grid(alpha=0.6)

axs[4].set_title('redshift')
axs[4].hist(full_df_cut['z'], bins=20)
axs[4].grid(alpha=0.6)
plt.savefig("user_distribution.jpg", format="jpg")
```

```python
for param in full_df_cut.columns:
    fig = plt.figure()
    plt.hist(full_df_cut[param], bins=99)
    plt.suptitle(param)
    plt.grid()
```

```python
# making galaxy pairs

n_pairs = 20000

columns=["zl", "l_mag_g", "l_mag_r", "l_mag_i", "l_mag_z", "zs", "s_mag_g", "s_mag_r", "s_mag_i", "s_mag_z"]

unified_df = pd.DataFrame(columns=columns)

denied_counter = 0 # para contar quantos foram negados no total

while unified_df.shape[0]<n_pairs:

    lens_index = np.random.randint(0, full_df_cut.shape[0])
    source_index = np.random.randint(0, full_df_cut.shape[0])

    if full_df_cut.iloc[lens_index, 0] < full_df_cut.iloc[source_index, 0]: # zl < zs

        #print("accepted")

        temp = pd.DataFrame([[
            full_df_cut.iloc[lens_index, 0],
            full_df_cut.iloc[lens_index, 1],
            full_df_cut.iloc[lens_index, 2],
            full_df_cut.iloc[lens_index, 3],
            full_df_cut.iloc[lens_index, 4],
            full_df_cut.iloc[source_index, 0],
            full_df_cut.iloc[source_index, 1],
            full_df_cut.iloc[source_index, 2],
            full_df_cut.iloc[source_index, 3],
            full_df_cut.iloc[source_index, 4],
        ]], columns=columns)
        unified_df = pd.concat([unified_df, temp], axis=0)
    else:
        denied_counter += 1

print(denied_counter + n_pairs)
```

```python
df_dict = {
    'zl':'CONFIGURATION_1-PLANE_1-OBJECT_1-REDSHIFT-g',
    'zs':'CONFIGURATION_1-PLANE_2-OBJECT_1-REDSHIFT-g',
    'thetaE':'PLANE_1-OBJECT_1-MASS_PROFILE_1-theta_E',
    'sig_v':'PLANE_1-OBJECT_1-MASS_PROFILE_1-sigma_v', # talvex tenha que colocar -g ou qualquer outra banda no final
    'l_effrad':'PLANE_1-OBJECT_1-LIGHT_PROFILE_1-R_sersic',
    'l_mag_g':'CONFIGURATION_1-PLANE_1-OBJECT_1-LIGHT_PROFILE_1-magnitude-g',
    'l_mag_r':'CONFIGURATION_1-PLANE_1-OBJECT_1-LIGHT_PROFILE_1-magnitude-r',
    'l_mag_i':'CONFIGURATION_1-PLANE_1-OBJECT_1-LIGHT_PROFILE_1-magnitude-i',
    'l_mag_z':'CONFIGURATION_1-PLANE_1-OBJECT_1-LIGHT_PROFILE_1-magnitude-z',
    'xs':'PLANE_2-OBJECT_1-LIGHT_PROFILE_1-center_x',
    'ys':'PLANE_2-OBJECT_1-LIGHT_PROFILE_1-center_y',
    'rs':'PLANE_2-OBJECT_1-LIGHT_PROFILE_1-R_sersic',
    's_mag_g':'CONFIGURATION_1-PLANE_2-OBJECT_1-LIGHT_PROFILE_1-magnitude-g',
    's_mag_r':'CONFIGURATION_1-PLANE_2-OBJECT_1-LIGHT_PROFILE_1-magnitude-r',
    's_mag_i':'CONFIGURATION_1-PLANE_2-OBJECT_1-LIGHT_PROFILE_1-magnitude-i',
    's_mag_z':'CONFIGURATION_1-PLANE_2-OBJECT_1-LIGHT_PROFILE_1-magnitude-z',
}

unified_df.rename(columns = df_dict, inplace = True)

display(unified_df.head())
```

```python
fig, axs = plt.subplots(2, 5, figsize=(15,6))

axs[0,0].set_title('lens g')
axs[0,0].hist(unified_df['CONFIGURATION_1-PLANE_1-OBJECT_1-LIGHT_PROFILE_1-magnitude-g'], bins=50)
axs[0,0].grid(alpha=0.6)

axs[0,1].set_title('lens r')
axs[0,1].hist(unified_df['CONFIGURATION_1-PLANE_1-OBJECT_1-LIGHT_PROFILE_1-magnitude-r'], bins=50)
axs[0,1].grid(alpha=0.6)

axs[0,2].set_title('lens i')
axs[0,2].hist(unified_df['CONFIGURATION_1-PLANE_1-OBJECT_1-LIGHT_PROFILE_1-magnitude-i'], bins=50)
axs[0,2].grid(alpha=0.6)

axs[0,3].set_title('lens z')
axs[0,3].hist(unified_df['CONFIGURATION_1-PLANE_1-OBJECT_1-LIGHT_PROFILE_1-magnitude-z'], bins=50)
axs[0,3].grid(alpha=0.6)

axs[0,4].set_title('source redshift')
axs[0,4].hist(unified_df['CONFIGURATION_1-PLANE_2-OBJECT_1-REDSHIFT-g'], bins=50)
axs[0,4].grid(alpha=0.6)

axs[1,0].set_title('source g')
axs[1,0].hist(unified_df['CONFIGURATION_1-PLANE_2-OBJECT_1-LIGHT_PROFILE_1-magnitude-g'], bins=50)
axs[1,0].grid(alpha=0.6)

axs[1,1].set_title('source r')
axs[1,1].hist(unified_df['CONFIGURATION_1-PLANE_2-OBJECT_1-LIGHT_PROFILE_1-magnitude-r'], bins=50)
axs[1,1].grid(alpha=0.6)

axs[1,2].set_title('source i')
axs[1,2].hist(unified_df['CONFIGURATION_1-PLANE_2-OBJECT_1-LIGHT_PROFILE_1-magnitude-i'], bins=50)
axs[1,2].grid(alpha=0.6)

axs[1,3].set_title('source z')
axs[1,3].hist(unified_df['CONFIGURATION_1-PLANE_2-OBJECT_1-LIGHT_PROFILE_1-magnitude-z'], bins=50)
axs[1,3].grid(alpha=0.6)

axs[1,4].set_title('source redshift')
axs[1,4].hist(unified_df['CONFIGURATION_1-PLANE_2-OBJECT_1-REDSHIFT-g'], bins=50)
axs[1,4].grid(alpha=0.6)

plt.tight_layout()
```
